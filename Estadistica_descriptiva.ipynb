#**Encuesta Anual de Hogares (EAH)**
####Dirección General de Estadística y Censos
Esta es una encuesta anual que realiza la DGESYC desde 2002. Se trata de un operativo por muestreo que involucra un número importante de viviendas particulares distribuidas en el territorio de la Ciudad. La muestra está diseñada de manera tal que los resultados de la encuesta permiten una representatividad del total de la Ciudad y de cada una de sus comunas.
De esta forma se propone recabar datos para conocer y analizar la situación socioeconómica y demográfica de la población y de los hogares de la Ciudad.
Conocer la situación del mercado laboral y el estado y la dinámica demográfica de la población es necesario para el diseño informado de la gestión pública.

##Distribución de la población por género y rango etario

```python
df = pd.DataFrame(datos)
intervalos_edad = list(range(18, 103, 12))
nombres_intervalos = [f'{i}-{i+11} años' for i in intervalos_edad[:-1]]
df['Intervalo de Edad'] = pd.cut(df['edad'], bins=intervalos_edad, labels=nombres_intervalos, right=False)
conteo = df.groupby(['Intervalo de Edad', 'sexo']).size().unstack(fill_value=0)
print(conteo)
```

#Graficos

###Histograma

```python
sns.histplot(x="edad", bins = 12, data=datos, kde=True, hue="sexo")
```

*Este Grafico muestra la cantidad de individuos separados por genero, agrupados por edades. Existe una concentración mayor de personas de aproximadamente 40 años. Es de tipo bimodal.*
___

```python
fig, axes = plt.subplots(ncols=2, figsize=(12,4))
axes = axes.flat
dt = ['ingreso_total_lab','ingreso_total_no_lab']
columnas_numeric = ['ingreso_total_lab', 'ingreso_total_no_lab']

for i, colum in enumerate(columnas_numeric):
    sns.histplot(
        data    = datos,
        x       = colum,
        stat    = "count",
        kde     = True,
        color   = (list(plt.rcParams['axes.prop_cycle'])*2)[i]["color"],
        line_kws= {'linewidth': 2},
        alpha   = 0.3,
        ax      = axes[i],
        hue     = "sexo",
        bins    = 40
    )
    axes[i].set_title(colum, fontsize = 10, fontweight = "bold")
    axes[i].set_ylabel('Cantidad')
    axes[i].tick_params(labelsize = 8)
    axes[i].set_xlabel("")

fig.tight_layout()
plt.subplots_adjust(top = 0.9)
fig.suptitle('Distribución variables numéricas', fontsize = 10, fontweight = "bold");
```

*Podemos observar mediante estos graficos, la cantidad de personas que registran determinados ingresos laborales y no laborales, categorizados por genero*

###Graficos de Torta

```python
datos_agrup=datos.groupby(['cat_ocupacional']).size().reset_index(name='Cantidad')

print(datos_agrup)

fig,ax=plt.subplots()
ax.pie(datos_agrup['Cantidad'], labels=["Asalariado","No Trabaja","Empleador","Trabajador familiar","Cuenta propia"], autopct="%0.1f %%")
ax.set_title("Cantidad de personas segun su estado ocupacional")
plt.show()
```

*Se puede observar que denotan dos categorias de grandes proporciones, el
**Asalariado** con 48,4% y **No trabaja** con el 33.1%.*
____

```python
datos_agrup=datos.groupby(['afiliacion_salud']).size().reset_index(name='Cantidad')

print(datos_agrup)

fig,ax=plt.subplots()
ax.pie(datos_agrup['Cantidad'], labels=["Otros","Obra Social","Prepaga","Mutual","Sist Publico"], autopct="%0.1f %%")
ax.set_title("Cantidad de personas con Afiliacion Sanitaria segun tipo")
plt.show()
```

*Mas del 50% de las personas entrevistadas cuentan con obra social*.
___

###Gráfico de dispersión

```python
g=sns.scatterplot(x='ingreso_total_no_lab', y='ingresos_totales', hue='sexo',data=datos)
g.set_title('Grafico 1:\nIngresos Totales en funcion de Ingresos no Laborales',y= 1.12, fontsize=10,fontweight='bold')
g.set_ylabel('Ingresos Totales')
g.set_xlabel('Ingresos no Laborales')
g.legend(loc="lower right")
plt.show()

g=sns.scatterplot(x='ingreso_total_lab', y='ingresos_totales', hue='sexo',data=datos)
g.set_title('Grafico 2:\nIngresos Totales en funcion de Ingresos Laborales',fontsize=10,fontweight='bold')
g.set_ylabel('Ingresos Totales')
g.set_xlabel('Ingresos Laborales')
g.legend(loc="lower right")
plt.show()
```

*El **Grafico 1** y **Grafico 2** muestran una correlacion fuerte con tendencia positva presentando un minimo de valores atipicos*.
___

```python
sns.pairplot(datos, x_vars=['ingreso_total_no_lab','ingreso_total_lab'], y_vars='ingresos_totales',hue='sexo', size=7, aspect=0.8,kind = 'reg')
```

*Comparando estos graficos, podemos deducir que la pendiente del **Grafico 2** es mayor que la del **Grafico 1**, es decir existe una mayor correlacion entre las variables involucradas.*

###BOX PLOT

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

datos_box=datos[datos['sexo']=='Varon']
datos_box=datos_box[['ingreso_total_lab','ingreso_total_no_lab','ingresos_totales']]
ax1.boxplot(datos_box)
ax1.set_title("Varones")

datos_box=datos[datos['sexo']=='Mujer']
datos_box=datos_box[['ingreso_total_lab','ingreso_total_no_lab','ingresos_totales']]
ax2.boxplot(datos_box)
ax2.set_title("Mujeres")
plt.show()
```

```python
datos_num=datos[['ingreso_total_lab','ingreso_total_no_lab','ingresos_totales']]

vals, names, xs = [],[],[]
for i, col in enumerate(datos_num.columns):
    vals.append(datos_num[col].values)
    names.append(col)
    xs.append(np.random.normal(i + 1, 0.04, datos_num[col].values.shape[0]))  # adds jitter to the data points - can be adjusted

plt.boxplot(vals, labels=names)
palette = ['r', 'g', 'b', 'y']
for x, val, c in zip(xs, vals, palette):
    plt.scatter(x, val, alpha=0.4, color=c)
plt.show()
```
